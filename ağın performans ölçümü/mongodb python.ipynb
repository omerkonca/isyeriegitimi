```python
import pymongo
import time

# MongoDB bağlantısı
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["mydatabase"]
collection = db["mycollection"]

# Veri alma işlemi
count = 0
start_time = time.time()
while True:
    document = collection.find_one()
    count += 1
    elapsed_time = time.time() - start_time
    if elapsed_time >= 1:
        print(f"Geçen süre: {elapsed_time:.2f} saniye, Alınan veri sayısı: {count}")
        count = 0
        start_time = time.time()
```
