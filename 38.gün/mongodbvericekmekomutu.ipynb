```python
import pymongo
from datetime import datetime
import random
import time

start_time = time.time()



for i in range(1000000):
    random.randint(1, 100)

end_time = time.time()

# MongoDB'ye balan
client = pymongo.MongoClient("mongodb://192.168.2.107:27017/")

# "test" veritabanna eri
db = client["robutel_local_log"]

# "orders" koleksiyonuna eri
collection = db["CurrentSpeedLog"]

# Filtreleme sorgusu
query = {
    "Timestamp": {
        "$gte": datetime(2023, 3, 1),
        "$lte": datetime(2023, 4, 10)
    }
}

# Filtrelenmi verileri getir
results = collection.find(query)
# Veri saysn al


# Her bir sonucu ekrana yazdr
for result in results:
    print(result)
    
 
print("Kodun alma sresi: ", end_time - start_time, " saniye")    
```
